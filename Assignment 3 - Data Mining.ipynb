7. **Implement a function that receives a DataFrame of frequent itemsets and a strong association rule (represented by a frozenset of antecedents and a frozenset of consequents). This function should return the corresponding Kulczynski measure. Include the code in your report**. 

```
from mlxtend.frequent_patterns import association_rules

strong_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
display(strong_rules)
```

```
def Kul_measure(a,b,c):
  km=((a / b) + (a / c) )/ 2
  return km
```

```
# [0, 1.0], higher is better
support = strong_rules['support']
cons_support = strong_rules['consequent support']
ante_support = strong_rules['antecedent support']
Kul_measure(support,cons_support,ante_support)
```
