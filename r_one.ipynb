```python

import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error,mean_absolute_error

def plot_regression_line(x, y, b): 
  # plotting the actual points as scatter plot 
  plt.scatter(x, y, color = "m", 
      marker = "o", s = 30) 

  # predicted response vector 
  y_pred = b[0] + b[1]*x  
  
  print('The model is:\n {}\n---'.format('b[0] + b[1]*x'))
  # plotting the regression line 
  plt.plot(x, y_pred, color = "g") 
  print('MSE: ',mean_squared_error(y,y_pred)) 
  print('MAE: ',mean_absolute_error(y,y_pred))
  
  # putting labels 
  plt.xlabel('x') 
  plt.ylabel('y') 

  # function to show plot 
  plt.show() 


def main():
 b_0 = np.array([1,0,-1,-0.5,0.5])
 b_1 = np.array([2,1,2,1,1])
 x = np.array([2,3,1,1,0,5,4,6,5,3]) 
 y = np.array([1,2,2,1,1,3,3,7,6,5])

 for intercept,x1 in np.nditer([b_0,b_1]):
    #print(intercept,x1)
    b0 = intercept
    b1 = x1 
    b = b0,b1
    # estimating coefficients
    print("Estimated coefficients:\nb_0 = {}  \nb_1 = {}".format(b0, b1))
    # plotting regression line 
    plot_regression_line(x, y, b)
   
if __name__ == "__main__": 
  main()     
```
