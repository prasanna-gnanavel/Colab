# 5. Function

We can write our own functions using the keyboard 'def' General syntax:

def <span style="color:blue">function_name</span>(parameters):

    body of the function
    
    return sth
    
Example: finding the sum of elements in a list using for loop
   


```python
def sum_list(data):
    """ Returns the sum of a list of numbers""" #Good habit to add comments
    sum = 0
    for x in data:
        sum+=x
    return sum
```

```python
#Using the above function
result = sum_list([1,2,3,4])
print(result)
```

# 6. Functions: returning multiple values

```python
def max_min(data):
    """ Returns min, max from the list"""
    min, max = data[0], data[0]
    for item in data:
        if item < min:
            min = item
        elif item > max:
            max = item
    return max, min
#This function returns two valus that will be returned as a tuple
#Usage:
            
```

```python
max, min = max_min([1,7,2,9,12])
#print('Max is '+ max+ ' and min is ' + min)
```

```python
print('Max is '+ str(max)+ ' and min is ' + str(min))
```

# 7. Functions: named and default parameters


### In function call, parameters can be explicitly named and we can initialize them with default values
### Default and named parameters can also be conveniently combined

```python
# Named parameters
```

```python
def compute_sum(a,b):
    return a + b
#Usages:
ans = compute_sum(10, 20)
ans = compute_sum(b=20, a=10)
```

```python
# Named + default parameters
def compute_sum(a, b, addC=False, c=99):
    if addC:
        return a + b + c
    else:
        return a + b
#Usage:
ans = compute_sum(10, 20, True)
```

```python
ans = compute_sum(a=10, b=20, True, 200) 
# be careful about the order of parameters to avoid ambiguilty
```

```python
ans = compute_sum(20, 10, True, 200)
```

# 24. File reading/writing

## General syntax:

### <span style="color:blue">f = open(filename, 'mode')</style>

## Example

```python
input_file = open('input.txt', 'r')
output_file = open('test.txt', 'w') 
for line in input_file:
    output_file.write(line)
    print(line)
input_file.close()
output_file.close()
```

## 'w' = 'writing mode'
## 'a' = 'append mode'
## 'wb' = 'write in binary'
## 'r' = 'read mode' (default)
## 'rb' = 'read in binary

### <span style="color:blue">Remember to close the file in the end</style>
