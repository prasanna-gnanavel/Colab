## 3. Basic data visualisation using Python


Data visualization is the display of information in a graphic or tabular format. Successful visualization requires that the data (information) be converted into a visual format so that the characteristics of the data and the relationships among data items or features can be analyzed or reported. 


---
> To display the histogram for the milieage feature by discretizing it into 8 separate bins and counting the frequency for each bin, run this code:

**<font color="red"></font>**

```python
%matplotlib inline

df['Engine'].hist(bins=8)
```

> A boxplot can also be used to show the distribution of values for each feature.

```python
df.boxplot()
```
