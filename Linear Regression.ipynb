```
#Question 1
import numpy as np

#change value here 
x = ([-7,-4,1,6,12,20,30,8,1,-5]) 
y = ([20,22,19,16,16,15,16,15,15,17])

order = 1
p = np.poly1d(np.polyfit(x, y, order))

#change value here 
t = np.linspace(-10, 50, 201)


plt.plot(x, y, 'o', label="")
#plt.plot(xTest, yTest, '*', label="Test Data")
plt.plot(t, p(t), '-', label="", linewidth=2)

#change value here 
plt.xlabel("Temperature", fontsize=18)
plt.ylabel("Abs", fontsize=18)
plt.xlim(-10,32)
plt.ylim(10,30)





# plt.gca().set_aspect('equal')
plt.grid(alpha=0.2)
plt.legend(fontsize=12)
plt.show()

print('The learned model is:\n {}\n---'.format(p))
    
MSE_Train = sum((p(x) - y)**2 for x, y in zip(x, y))/len(y)
print('Train MSE: {:8.5f}'.format(MSE_Train))

#MSE_Test = sum((p(x) - y)**2 for x, y in zip(xTest, yTest))/len(yTest)
#print('Test MSE: {:8.5f}'.format(MSE_Test))
```
